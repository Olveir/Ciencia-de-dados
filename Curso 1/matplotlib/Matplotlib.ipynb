```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.title('Valores do Combustivel em USD')

plt.plot(gas['Year'], gas['Australia'], 'y.-', label = 'Australia')
plt.plot(gas['Year'], gas['Italy'], 'g.-', label = 'Italy')
plt.plot(gas['Year'], gas['USA'], 'b.-', label = 'USA')
plt.plot(gas['Year'], gas['Japan'], 'r.-', label = 'Japan')

plt.xticks(gas['Year'][::2])

plt.xlabel('Anos')
plt.ylabel('Valor em USD')

plt.legend()
plt.show()
```

```python
plt.title('Valores do Combustivel em USD')

list_country = ['Australia', 'USA', 'Canada']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label = country, marker = '.')
        
plt.xticks(gas['Year'][::2])

plt.xlabel('Anos')
plt.ylabel('Valor em USD')

plt.legend()
plt.savefig('GraficoValcombust.pdf')#savefig tem que ser colocado antes do show.
plt.show()
```
